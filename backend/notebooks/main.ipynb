```python
import pandas as pd

from src.data_loader import load_data, BASE_DIR
from src.preprocess import clean_data, validate_schema

# Load data
df, config = load_data()

# Validate schema
validate_schema(df)

# Basic info
print("Shape:", df.shape)
print("\nNull values:\n", df.isnull().sum())
print("\nData types:\n", df.dtypes)

# Clean data
df_clean = clean_data(df)

print("\nAfter Cleaning Shape:", df_clean.shape)

# Validate output config
if "output" not in config or "path" not in config["output"]:
    raise ValueError("Output path missing in config")

output_path = BASE_DIR / config["output"]["path"]


output_path.parent.mkdir(parents=True, exist_ok=True)

# Save
df_clean.to_csv(output_path, index=False)

print("\nSaved to:", output_path)

df_clean.head()
```
